# Linear Regression Explained

Linear regression is a statistical method used to model and analyze the relationship between a dependent variable (y) and one or more independent variables (x).

## Linear regression fomula

$$
Y = b_0 + b_1X + \epsilon
$$

- ***Y*** dependent variable
- ***X*** independent variable
- ***b0*** intercept of the line (the value of Y when X is 0)
- ***b1*** is the slope of the line (the change in Y for a one-unit change in X)
- ***e*** is the error term (the difference between the observed and predicted values)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([-1, 0, 1, 2, 3, 4, 5, 6])

y1_b0 = 2
y1_b1 = 3
y1 = y1_b0 + y1_b1 * x

y2_b0 = 5
y2_b1 = 10
y2 = y2_b0 + y2_b1 * x

y3_b0 = 3
y3_b1 = -10
y3 = y3_b0 + y3_b1 * x

plt.figure(figsize=(5, 3))
plt.plot(x, y1, label=f'Y = {y1_b0} + {y1_b1} * X', marker='o')
plt.plot(x, y2, label=f'Y = {y2_b0} + {y2_b1} * X', marker='o')
plt.plot(x, y3, label=f'Y = {y2_b0} + {y2_b1} * X', marker='o')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Line')
plt.grid(True)
plt.legend()
plt.show()
```
